```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Instagram_Analytics - final.csv")
df
```

```python
performance_summary = df[['likes','comments','shares','saves','reach','impressions']].describe()
performance_summary
```

```python
#overall average engagement rate
df['engagement_rate'].mean()
```

```python
media_engagement = df.groupby('media_type')['engagement_rate'].mean().reset_index()
media_engagement
```

```python
plt.figure()
sns.barplot(data=media_engagement , x='media_type' , y='engagement_rate')
plt.title("Average Engagement Rate by Media Type")
plt.show()
```

```python
content_engagement=df.groupby('content_category')['engagement_rate'].mean().reset_index()
content_engagement
```

```python
plt.figure()
sns.piechart(data=df, x='content_category', y='engagement_rate')
plt.xticks(rotation=45)
plt.title("Engagement Rate by Content Category")
plt.show()
```

```python
plt.figure()
plt.pie(
    media_engagement['engagement_rate'],
    labels=media_engagement['media_type'],
    autopct='%1.1f%%',
    startangle=90
)
plt.title("Average Engagement Rate by Media Type")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df, x='caption_length', y='engagement_rate')
plt.title("Caption length and Engagement Rate")
plt.show()
```

```python
plt.figure()
sns.scatterplot(data=df, x='hashtags_count', y='reach',)
plt.title("Hashtags count vs Reach")
plt.show()
```

```python
traffic_analysis = df.groupby('traffic_source')[['engagement_rate','followers_gained']].mean().reset_index()
traffic_analysis                                                                                        
```

```python
traffic_analysis.plot(kind='bar',x='traffic_source')
plt.title("Traffic source vs Engagement & Followers Gained")
plt.show()
```

```python
Insights:

Reels and videos show higher engagement rates than static images.
Medium-length captions generate better engagement.
Posts with moderate hashtag counts achieve higher reach.
Certain traffic sources contribute more to follower growth.
```
