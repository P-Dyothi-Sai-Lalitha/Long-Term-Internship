```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, norm
```

```python
data=pd.read_csv("Iris.csv")
data
```

```python
s_len = data["SepalLengthCm"]
s_len
```

```python
s_len = data["SepalLengthCm"]
print(s_len)
a1=s_len.mean()
b1=s_len.std()
c1=s_len.var()
d1=s_len.median
print(a1)
print(b1)
print(c1)
print(d1)
```

```python
s_wid = data["SepalWidthCm"]
print(s_wid)
a2=s_wid.mean()
b2=s_wid.std()
c2=s_wid.var()
d2=s_wid.median()
print(a2)
print(b2)
print(c2)
print(d2)
```

```python
p_len = data["PetalLengthCm"]
print(p_len)
a3=p_len.mean()
b3=p_len.std()
c3=p_len.var()
d3=p_len.median()
print(a3)
print(b3)
print(c3)
print(d3)
```

```python
p_wid = data["PetalWidthCm"]
print(p_wid)
a4=p_wid.mean()
b4=p_wid.std()
c4=p_wid.var()
d4=p_wid.median()
print(a4)
print(b4)
print(c4)
print(d4)
```

```python
p=s_len.isnull().sum()
q=s_wid.isnull().sum()
r=p_len.isnull().sum()
s=p_wid.isnull().sum()
print(p)
print(q)
print(r)
print(s)
```
