```python
import numpy as np
print("Vectorization is used to remove the use of loops")
arr = np.array([1,2,3,4,5])
number = 10
result = arr + number
print(result)
```

```python
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[0,1,2],[0,1,2]])
arr_sum = arr1+arr2
print(arr_sum)
```

```python
array = np.array([-1,0,1,2,3])
def find_Square(x):
    if x<0:
        return 0
    else :
        return x**2
vectorized_function = np.vectorize(find_Square)
result = vectorized_function(array)
print(result)
```

```python
print("Broadcasting Arrays")
array1 = np.array([1,2,3])
array2 = np.array([[1],[2],[3]])
result = array2+array1
print(result)
```
