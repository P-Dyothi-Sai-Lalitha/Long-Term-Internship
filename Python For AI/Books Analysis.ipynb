```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stat
```

```python
data = pd.read_csv("Books.csv")
data
```

```python
data.isnull().sum()
```

```python
data.duplicated().sum()
```

```python
rating = data["rating_average"]
rating
```

```python
r1=rating.mean()
r2=rating.median()
r3=rating.std()
r4=rating.var()
print(r1)
print(r2)
print(r3)
print(r4)
```

```python
import scipy
from scipy.stats import norm
rating.hist(figsize=(8,6), density=True, bins=10) 
x=np.linspace(rating.min(), rating.max(), 100)
y=norm.pdf(x,r1,r3)
plt.plot(x,y)
plt.show
```
