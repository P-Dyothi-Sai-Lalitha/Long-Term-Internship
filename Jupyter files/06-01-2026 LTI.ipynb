```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.metrics import log_loss, roc_auc_score
import numpy as np
```

```python
y_true=np.array([1,0,1,0,1,0,1,0])
y_pred_proba=np.array([0.9,0.2,0.8,0.4,0.7,0.1,0.6,0.3])
y_pred_class=(y_pred_proba>=0.5).astype(int)
```

```python
precision=precision_score(y_true,y_pred_class)
print(f"precision: {precision: .3f}")
```

```python
recall=recall_score(y_true,y_pred_class)
print(f"recall: {recall: .3f}")
```

```python
f1=f1_score(y_true,y_pred_class)
print(f"f1_score: {f1: .3f}")
```

```python
logloss=log_loss(y_true,y_pred_class)
print(f"log_loss: {logloss: .3f}")
```

```python
auc=roc_auc_score(y_true,y_pred_class)
print(f"roc_auc_score: {auc: .3f}")
```
